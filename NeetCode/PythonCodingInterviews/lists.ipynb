Python by default uses resizable lists. Allowing for elements to be added or removed.
pop() removes and returns the last element of a list. pop can also be used to remove a specific index
append() this will add an element to the end of a list
insert(1,3) will add to after a specific index

Challenge

Implement the following functions:

    append_elements(arr1: List[int], arr2: List[int]) -> List[int]. It should append the elements of arr2 to the end of arr1 and return the resulting list.
    pop_n(arr: List[int], n: int) -> List[int]. It should remove the last n elements from the list and return the resulting list. If n is greater than the length of the list, it should return an empty list.
    insert_at(arr: List[int], index: int, element: int) -> List[int]. It should insert the element at the specified index in the list and return the resulting list. If the index is out of bounds, you should insert it at the end of the list.


```python
from typing import List


def append_elements(arr1: List[int], arr2: List[int]) -> List[int]:
    for i in arr2:
        arr1.append(i)
    return arr1


def pop_n(arr: List[int], n: int) -> List[int]:

    while n > 0 and arr:
        arr.pop()
        n -= 1

    return arr

def insert_at(arr: List[int], index: int, element: int) -> List[int]:
    if len(arr) <= index:
        arr.append(element)
    else:
        arr.insert(index,element)
    return arr


# do not modify below this line
print(append_elements([1, 2, 3], [4, 5, 6]))
print(append_elements([4, 3], [4, 5, 3]))

print(pop_n([1, 2, 3, 4, 5], 2))
print(pop_n([1, 2, 3, 4, 5], 6))
print(pop_n([1, 2, 3, 4, 5], 5))

print(insert_at([1, 2, 3, 4, 5], 2, 6))
print(insert_at([1, 2, 3, 4], 6, 5))
```

Challenge

Implement the following functions:

    append_elements(arr1: List[int], arr2: List[int]) -> List[int]. It should append the elements of arr2 to the end of arr1 and return the resulting list. Yes, this is the same function from the previous lesson.
    remove_elements(arr1: List[int], arr2: List[int]) -> List[int]. It should remove all elements of arr2 from arr1 and return the resulting list.
        If any of the elements in arr2 are not in arr1, then skip them.
        You don't have to worry about implementing an efficient solution. Try to use the built-in functions.


```python
from typing import List


def append_elements(arr1: List[int], arr2: List[int]) -> List[int]:
    arr1.extend(arr2)
    return arr1
  

def remove_elements(arr1: List[int], arr2: List[int]) -> List[int]:
    for num in arr2:
        if num in arr1:
            arr1.remove(num)
    return arr1


# do not modify below this line
print(append_elements([1, 2, 3], [4, 5, 6]))
print(append_elements([4, 3], [4, 5, 3]))

print(remove_elements([1, 2, 3, 4, 5], [2, 4, 6]))
print(remove_elements([1, 2, 3, 4, 5], [2, 3, 4, 5, 5]))
print(remove_elements([1, 7, 2, 3, 4, 5], [6, 7, 8, 2]))
```

If we want to create an empty list of a specific size, this is the easiest way to do it in Python

my_list = [0] * 5

The above code will create a list of 5 zeros. It might seem strange to multiply a list by a number, but this is the standard way to create a list of a specific size in Python. We could have replaced the 0 with any other value.

my_list = [1] * 3

The above code will create a list of 3 ones.

Challenge

Implement the following function:

    create_list_with_value(size: int, index: int, value: int) -> List[int] which should create and return a list of length size. All values in the list should be 0, except for the value at index index, which should be the parameter value.
        You may assume the index is within the bounds of the list.

Time Complexity

    Time complexity: O(n)O(n) where nn is the size of the resulting list.


```python
from typing import List


def create_list_with_value(size: int, index: int, value: int) -> List[int]:
    my_list = [0] * size
    my_list[index] = value
    return my_list



# do not modify below this line
print(create_list_with_value(5, 3, 7))
print(create_list_with_value(1, 0, 5))
print(create_list_with_value(10, 9, 9))
print(create_list_with_value(10, 9, 0))
```

Three methods to clone a list;

    Using the copy() method:

original_list = [1, 2, 3]
cloned_list = original_list.copy()

    Using the slicing syntax:

original_list = [1, 2, 3]
cloned_list = original_list[:]

    Using the list() constructor:

original_list = [1, 2, 3]
cloned_list = list(original_list)

This does not create a deepcopy of the list. Instead we have to use the code below
import copy

original_list = [[1, 2], [3, 4]]

cloned_list = copy.deepcopy(original_list)

    remove_element(arr: List[int], element: int) -> List[int] that takes a list of integers and returns a new list with the specified element removed. You should not modify the original list.
        You may assume the element is always in the list.

Tip: It's okay if you don't remember how to remove an element from a list. You can look back at a previous lesson or do a quick search to refresh your memory.
Time Complexity

    Time complexity: O(n)O(n) where nn is the length of the list being copied.



```python
from typing import List


def remove_element(arr: List[int], element: int) -> List[int]:
    new_list = list(arr)
    new_list.remove(element)

    return new_list



# do not modify below this line
arr = [1, 3, 5, 7, 9]

print(remove_element(arr, 3))
print(arr)
print(remove_element(arr, 9))
print(arr)
print(remove_element(arr, 1))
print(arr)
```

List Comprehension

A very powerful feature in Python is comprehension. It applies to lists and other data types in Python that we will cover later. It allows us to create lists in a concise way. Here's an example:

my_list = [i for i in range(5)]

print(my_list) # [0, 1, 2, 3, 4]

This is called list comprehension. In the above example, we have a for loop inside the square brackets. The loop iterates over the range of numbers from 0 to 4. Before the loop we place the expression i which is the value we want to add to the list. This is a very concise way to create a list.

And it's very flexible:

arr1 = [1, 2, 3]
arr2 = [4, 5, 6]

result = [i + j for i, j in zip(arr1, arr2)]

Above we have two lists arr1 and arr2. We use the zip() function to iterate over both lists at the same time. We then add the elements of both lists together and create a new list. The expression i + j is the value we want to add to the list.

We can also add a condition to the comprehension:

arr = [1, 2, 3, 4, 5]

result = [i for i in arr if i % 2 == 0]

In the above example, we have a list arr. We iterate over the list and add the element to the new list only if it's even. This might seem like overkill since we can accomplish the same thing with a simple loop, but it demonstrates the flexibility of list comprehension.

Challenge

Implement the following function using list comprehension:

    create_list_of_odds(n: int) -> List[int] which takes an integer n and returns a list of all odd numbers from 1 to n (inclusive).

Hint: You can use the range() function and pass three arguments to it: the start (inclusive), the end (non-inclusive), and the step.
Time Complexity

    Time complexity: O(n)O(n) where nn is the number of times the loop within the list comprehension runs.


```python
from typing import List


def create_list_of_odds(n: int) -> List[int]:
    return [i for i in range(1, n+1,2)]


# do not modify below this line
print(create_list_of_odds(1))
print(create_list_of_odds(5))
print(create_list_of_odds(6))
print(create_list_of_odds(10))
```
